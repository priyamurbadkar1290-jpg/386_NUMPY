# NumPy and Math

```python
import numpy as np
```

```python
python_list = [1, 2, 3, 4, 5]

numpy_array = np.array(python_list)
print(numpy_array)
```

```python
oned_shape = numpy_array.shape
oned_size = numpy_array.size
oned_type = numpy_array.dtype

print('1D NumPy Array:')
print("Array:", numpy_array)
print(f'Shape: {oned_shape}')
print(f'Size: {oned_size}')
print(f'DataType: {oned_type}')
```

```python
arr1 = np.array([10, 20, 30, 40])
arr2 = np.array([2, 4, 6, 8])
```

```python
addition = arr1 + arr2
print(addition)
```

```python
subtraction = arr1 - arr2
print(subtraction)
```

```python
multiplication = arr1 * arr2
print(multiplication)
print(multiplication.dtype)
```

```python
division = arr1 / arr2
print(division)
print(division.dtype)
```

```python
data = np.array([85, 90, 78, 92, 88, 85, 95])
```

```python
mean = np.mean(data)
median = np.median(data)
minimum = np.min(data)
std_dev = np.std(data)
variance = np.var(data)
range_value = np.max(data) -np.min(data)

print('Mean:', mean)
print('Median:', median)
print('Min:', minimum)
print('Std Dev:', std_dev)
print('Variance:', variance)
print('Range:', range_value)
```

```python
temp_data = np.array([22.3, 23.1, 24.5, 25.8, 23.6, 26.7, 27.9, 29.2, 30.5, 24.7, 23.4, 22.1, 25.3, 26.4, 28.7, 29.8, 31.2, 32.4, 30.7, 29.5, 27.8, 26.6, 23.9, 22.5, 24.1, 25.7, 27.3, 29.6, 31.0, 33.1, 31.9])
```

```python
mean_temp =  np.mean(temp_data)
median_temp = np.median(temp_data)
std_dev_temp = np.std(temp_data)

print(f'Mean: {mean_temp}')
print(f'Median: {median_temp}')
print(f'Std Dev: {std_dev_temp}')
```

```python
temp_in_f = (temp_data * 9/5) + 32
print(temp_in_f)
```

```python
hot_days = temp_data[temp_data > 30]
print('Hot Temp Days:', hot_days)
print('Num of Hot Days:', len(hot_days))
```

```python
base_temp = 20
cooling_degrees_total = np.sum(np.maximum(temp_data - base_temp,0))

print('Cooling Degrees:', cooling_degrees_total)
```
