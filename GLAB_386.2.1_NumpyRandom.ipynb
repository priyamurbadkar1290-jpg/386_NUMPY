# GLAB 386.2.1 Numpy Random

```python
from numpy import random
```

### Random module Functions
-   random.rand()- creates random  nums between 0 - 1
-   random.randn()- generate a single  random number
-   random.randn(n)- generates an array of n random  numbers
-   randon.rand(m, n)- generates a 2d  array  with m rows  and n columns
-   random.choice()
-   random.shuffle()- shuffle contents  of array



```python
data =  random.rand(10)

print(data)
```

```python
random_numbers = random.randn(5)

print(random_numbers)
```

```python
elements = ['a', 'b', 'c', 'd', 'e', 'f']
random_sample = random.choice(elements)

print(random_sample)
```

```python
random.shuffle(elements)
print(elements)
```

```python
# Define a list of dishes on the menu
menu = [
    "Spaghetti Carbonara",
    "Chicken Alfredo",
    "Margherita Pizza",
    "Cheeseburger",
    "Caesar Salad",
    "Fish and Chips",
    "Pad Thai",
    "Sushi Platter",
    "Vegetable Stir-Fry",
    "Grilled Salmon"
]
```
