```python
# arange to create quick numerical arrays (start, stop, step)
one_to_100 = np.arange(1, 100)

print(one_to_100)
```

```python
# A range array range 
arr = np.arange(1, 11)
print(arr)
```

```python
# Arithmetic Operations Numpy
# Element-wise arithmatic (for each element)
sum_arr = arr + 10
print(sum_arr)
```

```python
sub_arr = arr - 4
print(sub_arr)
```

```python
div_arr = arr / 2
print(div_arr)
```

```python
mult_arr = arr * 100
print(mult_arr)
```

```python
sum = np.sum(arr)
mean = np.mean(arr)
median = np.median(arr)
variance = np.var(arr)
standard_dev = np.std(arr)
min = np.min(arr)
max = np.max(arr)

print('Sum:', sum)
print('Mean:', mean)
print('Median:', median)
print(f'Standard dev: {standard_dev}')
print(f'Variance: {variance}')
print(f'Min: {min}')
print(f'Max: {max}')
```

```python
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix)
```

```python
transposed_matrix = matrix.T
print(transposed_matrix)
```

```python
reshaped_matrix = matrix.reshape(3, 2)
print(reshaped_matrix)
```

```python
# argmin(), arg(max) - index of min and max values
print(np.argmin(arr))
print(np.argmax(arr))
```

```python
# argmin(), arg(max) - index of min and max values
print(np.argmin(arr))
print(np.argmax(arr))
```
